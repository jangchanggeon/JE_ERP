```python
import pandas as pd
```

```python
def read_xlsx_dataframe(filename,sheetname):
    df = pd.read_excel(filename, engine='openpyxl',sheet_name=sheetname)
    data_frame_df = pd.DataFrame(df)

    # print(pd.DataFrame(data_frame_df.set_index('품번').T.to_dict()))
    if sheetname=='GLOVIS발주':
        dict__dataframe = pd.DataFrame(data_frame_df.set_index('부품번호').T.to_dict())
    elif sheetname=='덕양실적':
        dict__dataframe = pd.DataFrame(data_frame_df.set_index('품목코드').T.to_dict())
    elif sheetname=='LIST':
        dict__dataframe = pd.DataFrame(data_frame_df.set_index('서브품번').T.to_dict())
    else:
        dict__dataframe = pd.DataFrame(data_frame_df.set_index('품번').T.to_dict())
    
    return dict__dataframe
```

```python
import pandas as pd

append_dday = {}
def orderdata_dict(order_company):
#덕양scm

    if order_company=='덕양':
        append_dday = {}
        for k,v in order_df_duk.iteritems():
            if type(k) is str:
                if k in append_dday.keys():
                    append_dday[k] += v[5:17]
                else:
                    append_dday[k] = v[5:17]
                append_dday[k].loc['발주'] = [order_company]
        return append_dday
                
# 미창서열        
    elif order_company=='미창': 
        append_dday = {}
        for k,v in order_df_mch.iteritems():
            if type(k) is str:
                if k in append_dday.keys():
                    append_dday[k] += v[4:18]
                else:
                    append_dday[k] = v[4:18]
                append_dday[k].loc['발주'] = [order_company]      
        return append_dday
#HMC
    elif order_company=='HMC':
        append_dday = {}
        for k,v in order_df_hmc.iteritems():
            if type(k) is str:
                if k in append_dday.keys():
                    append_dday[k] += v[4:17]
                else:
                    append_dday[k] = v[4:17]
                append_dday[k].loc['발주'] = [order_company]
        return append_dday
#hmc    
    elif order_company=='hmc': 
        append_dday = {}
        for k,v in order_df_hmc.iteritems():
            if type(k) is str:
                if k in append_dday.keys():
                    append_dday[k] += v[4:17]
                else:
                    append_dday[k] = v[4:17]
                append_dday[k].loc['발주'] = [order_company]
        return append_dday

    else:
        print("해당시트가 없습니다.")
# ITEMLIST
#하위품번 : 완성품품번,고객사, 차종, CAV, U/S, C/T, 중량, GATE중량
def itemlist_dict():
    itemlist_df = {}
    for k,v in item_df.iteritems():
        if type(k)!=float:
            itemlist_df[k]={'품번':v['품번'],'고객사':v['고객사'],'차종':v['차종'],'CAV':v['CAV'],'U/S':v['U/S'],'C/T':v['C/T'],'중량':v['중량'],'GATE중량':v['GATE중량']}
    return itemlist_df
```
